```python
train_DIR = r"E:\pos_tag\pos_tag\corpus\treebank\train.txt"
test_DIR = r"E:\pos_tag\pos_tag\corpus\treebank\test.txt"
```

```python
data_train = pd.read_csv(train_DIR,delimiter="\t",names=("Từ","Loại từ"),quoting=csv.QUOTE_NONE, encoding='utf-8')
data_test = pd.read_csv(test_DIR,delimiter="\t",names=("Từ","Loại từ"),quoting=csv.QUOTE_NONE, encoding='utf-8')
```

```python
tagset = []
for tag in data_train["Loại từ"]:
    tagset.append(tag)
for tag in data_test["Loại từ"]:
    tagset.append(tag)
```

```python
tagset = list(set(tagset))
```

```python
print(len(tagset))
print(tagset)
```

```python
count_Tag = {}
for tag in tagset:
    count_Tag[tag] = 0
```

```python
for i in tagset:
    for j in data_train["Loại từ"]:
        if(i == j):
            count_Tag[i] +=1
```

```python
def count_tag1_tag2(tag1,tag2):
    count = 0
    for i in range(0,len(data_train["Loại từ"])-1):
        if(data_train["Loại từ"][i] == tag1):
            if(data_train["Loại từ"][i+1] == tag2):
                count +=1
    return count
        
```

```python
tags_matrix = np.zeros((len(tagset),len(tagset)),dtype='float64')
```

```python
for i,tag1 in enumerate(tagset):
    for j,tag2 in enumerate(tagset):
        tags_matrix[i,j] = count_tag1_tag2(tag1,tag2)/count_Tag[tag1]
```

```python
tags_matrix = pd.DataFrame(tags_matrix,columns=tagset,index=tagset)
```

```python
tags_matrix
```

```python
def count_word_tag(word,tag):
    count = 0
    for i in range(0,len(data_train)):
        if word == data_train["Từ"][i]:
            if tag == data_train["Loại từ"][i]:
                count += 1
    return count
        
```

```python
def Viterbi(words):
    state = []
    for key,word in enumerate(words):
        p = []
        for tag in tagset:
            if(key == 0):
                transition_p = tags_matrix.loc['CH',tag]
            else:
                transition_p = tags_matrix.loc[state[-1],tag]
            p_word_tag = count_word_tag(word,tag)/count_Tag[tag]
            state_probability = p_word_tag * transition_p
            p.append(state_probability)
            
        pmax = max(p)
        state_max = tagset[p.index(pmax)]
        state.append(state_max)
        print("{} : {}".format(word,state_max))
    return list(zip(words,state))
```

```python
tagged_word = Viterbi(test)
```

```python
#accuracy
count_acc = 0
for i in range(0,len(data_test["Loại từ"])):
    if(tagged_word[i][i] == data_test["Loại từ"][i]):
        count_acc +=1
accuracy = count_acc/len(data_test["Loại từ"])
print(accuracy)
```
